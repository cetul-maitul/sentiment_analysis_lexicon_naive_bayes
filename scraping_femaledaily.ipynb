```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import time

# List untuk menyimpan hasil scraping
reviews = []
usernames = []
ratings = []

# Loop sesuai jumlah halaman yang ingin diambil
for page in range(1, 150):  # Ubah 11 jadi jumlah halaman + 1
    url = f'https://reviews.femaledaily.com/products/treatment/skin-soothing-treatment/nature-republic/aloe-vera-92-soothing-gel-06?cat=&cat_id=0&age_range=&skin_type=&skin_tone=&skin_undertone=&hair_texture=&hair_type=&order=newest&page={page}'
    print(f"Scraping halaman {page}...")

    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    html = soup.find_all('div', {"class": "review-card"})

    if not html:
        print("Tidak ada review lagi.")
        break

    for row in html:
        review = row.find("p", {"class": "text-content"})
        username = row.find("p", {"class": "profile-username"})

        if review and username:
            reviews.append(review.text.strip())
            usernames.append(username.text.strip())

    time.sleep(1)  # Delay untuk menjaga agar tidak diblokir

# Buat DataFrame
df = pd.DataFrame({'Username': usernames, 'Reviews': reviews})
df
```

```python
df.to_csv('ulasan_femaledaily.csv', index=False, encoding='utf-8')
print("Scraping selesai. Data disimpan ke 'ulasan_femaledaily.csv'")
```
